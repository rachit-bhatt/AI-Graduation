## Model Building (KNN - K Nearest Neighbour):

### Getting data ready for the model.

<font color = 'lightblue'><h2>Syntax: iloc[from_row : to_row, from_col : to_col]</h2></font>

```python
# Rows - Independent Variables:
x = np.array(df.iloc[ : , : 4])
x
```

<font color = '#0F0'><h1>Hint/Trick/Tip: L for Label</h1></font>

```python
# Columns - `target` (Target Variable):
y = np.array(df.iloc[ : , 4 : ])
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
X_train, X_test, y_train, y_test
```

# KNN Model:

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Usually, we take square root of number of columns initially for `n_neighbors` for the first model.

```python
knn1 = KNeighborsClassifier(n_neighbors = 3)
knn1
```

```python
knn1.fit(X_train, y_train)
```

```python
y_pred = knn1.predict(X_test)
y_pred
```

## Model Accuracy:

```python
def display_model_accuracy(score):
    print(f'Model Accuracy: { int(score * 100) }%')
```

```python
#print(f'Model Accuracy: { int(knn1.score(X_test, y_pred) * 100) }%')
display_model_accuracy(knn1.score(X_test, y_pred))
```

### Confusion Matrix - as this is a classification model.

```python
from sklearn.metrics import accuracy_score
```

```python
display_model_accuracy(accuracy_score(y_test, y_pred))
```

## Confusion Matrix:

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred,
                 labels = [0, 1, 2])
```

## Classification Report:

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred,
                            labels = [0, 1, 2]))
```
