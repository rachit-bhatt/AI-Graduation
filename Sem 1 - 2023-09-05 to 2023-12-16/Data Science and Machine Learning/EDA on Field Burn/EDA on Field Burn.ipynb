## Q6. Ensure that the `date` columns have the correct datatype.

```python
# Finding the datatypes of the `date` fields.
date_fields = ['discovery_date', 'contain_date'] # Making a list of date fields available in the dataset for simplicity.
df[date_fields].dtypes # Checking their datatypes.
# Can also use info() method to explore some more information regarding the dataframe or series.
```

```python
# Checking the format of the dates in those fields.
df[date_fields].head(10)
```

```python
# Converting the given dates from `object (string)` to `datetime`.

# Method 1 - using Pandas.to_datetime() method.
for col in date_fields:
  # If the given date is not in proper format, this might generate an error.
  # In order to prevent it, we can provide an argument 'errors' to fix this.
  df[col] = pd.to_datetime(df[col], errors = 'coerce') # Converts `object` to `datetime64[ns]`.
```

```python
# Method 2 - using DataFrame.astype() method.
for col in date_fields:
  df[col] = df[col].astype('datetime64[ns]')
```

## Q8. Now explain whether you think the existing data is dependable or not, and whether you should/should not want to handle the missing data in `contain_date`.

<font color = '#0AA'>The `contain_date` determines how long did it took to handle the fire hazard happened in the specific <i>location</i>.

<font color = '#090'>As we can indentify from the above <u>Figure 7.1</u> and <u>Figure 7.2</u>, acres over 10000 having less missing values can be removed because the number is less.
<br>
However, if we impute to the whole column on basis of this data, the data below 10000 will be affected as well.
<br>
Hence, we should omit the records have `acres_burned` over 10000 and impute the missing values below 10000 `acres_burned` with the mean of the `contain_date` field.</font>

```python
# Omitting Missing Values:
df[df.acres_burned > 10000].contain_date.dropna(inplace = True)

# Imputing Missing Values:
df[df.acres_burned <= 10000].contain_date.fillna(value = df.contain_date.mean(), inplace = True)
```
