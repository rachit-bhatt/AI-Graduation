# 9-1. Restaurant:


```python
class Restaurant:
    '''
    Information and operations that are handled by a restaurant.
    '''

    def __init__(self, restaurant_name, cuisine_type) -> None:
        '''
        Stores the name and specialty of the restaurant.
        ---
        :param restaurant_name: Name of the Restaurant.
        :param cuisine_type: Specialty of the Restaurant.
        :type restaurant_name: str
        :type cuisine_type: str
        ---
        '''
        self.restaurant_name = restaurant_name.title()
        self.cuisine_type = cuisine_type

    def describe_restaurant(self) -> str:
        '''
        Describes the restaurant with its specialty.
        ---
        
        ---
        '''
        return f'{ self.restaurant_name }\'s { self.cuisine_type } is very delicious!'

    def open_restaurant(self) -> str:
        '''
        Describes whether the restaurant is open or not.
        ---
        
        ---
        '''
        return f'{ self.restaurant_name } is opened.' if random.random() < 0.5 else f'{ self.restaurant_name } is closed.' # Using Lambda Technique.
```

```python
#region Unit-Testing

if __name__ == '__main__':
    # Object Declaration and Initialization.
    mc_donald = Restaurant('McDonald', 'Burger')

    # Class Members.
    print(f'Restaurant Name: { mc_donald.restaurant_name }')
    print(f'Cuisine Type: { mc_donald.cuisine_type }')

    # Member Functions or Methods.
    print(mc_donald.describe_restaurant()) # Approach 1: object_name.method_name()
    print(Restaurant.open_restaurant(mc_donald)) # Appraoch 2: ClassName.method_name(object_name)

#endregion
```

# 9-3. Users:


```python
class User:
    '''
    A person interacting with our application.
    '''
    
    #region Attributes
    
    first_name = last_name = ''
    age = 0
    is_male = True # Gender

    #endregion

    def __init__(self, first_name = '', last_name = '', age = 0, is_male = True) -> None:
        '''
        Initializing the User with data such as name, age, and gender. Can act as a Default Constructor.
        ---
        
        ---
        '''
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.is_male = is_male

    def get_basic_gender(self) -> str:
        '''
        Provides the basic gender of the user based on the `is_male` attribute of the User initialized.
        ---
        
        ---
        '''
        return 'Male' if self.is_male else 'Female'

    def describe_user(self) -> str:
        '''
        Describes a summary of the information of the User.
        ---
        
        ---
        '''
        return f'User Summary:\nFirst Name: { self.first_name }\nLast Name: { self.last_name }\nAge: { self.age }\nGender: { self.get_basic_gender() }'

    def greet_user(self) -> str:
        '''
        Provides a personalized message for greeting the User.
        ---
        
        ---
        '''
        return f'Welcome to our application, { self.first_name }.'
```

```python
#region Unit-Testing

if __name__ == '__main__':
    # Creating multiple instances of the class User.
    rachit = User('Rachit', 'Bhatt', 23)
    manish = User('Manish', 'Malhotra', 56)
    john = User('John', 'Miller', 62)
    fatima = User('Fatima', 'Shaikh', 31, False)

    # Describing the User.
    print('User Description:')
    print('-----------------')
    print(f'\n{ rachit.describe_user() }')
    print(f'\n{ manish.describe_user() }')
    print(f'\n{ john.describe_user() }')
    print(f'\n{ fatima.describe_user() }')
    print('\n-----------------\n')

    # Greeting the User.
    print('Greeting Users:')
    print('---------------')
    print(f'\n{ rachit.greet_user() }')
    print(f'\n{ manish.greet_user() }')
    print(f'\n{ john.greet_user() }')
    print(f'\n{ fatima.greet_user() }')

#endregion
```

# Odometer:


```python
class Car:
    def __init__(self, make, model, year) -> None:
        '''
        Initialize attributes to describe a car.
        ---
        :param make: Maker/Manufacturer of the car.
        :param model: Model name of the car.
        :param year: Manufacturing year of the car.
        :type make: str
        :type model: str
        :type year: int
        :return: Since this is a constructor, it will not return anything.
        :rtype: None
        ---
        '''
        self.make = make.title()
        self.model = model.title()
        self.year = year
        self.odometer_reading = 0

    def get_descriptive_name(self) -> str:
        '''
        Describes the details of the car.
        ---
        :return: Descriptive text describing the details of the car.
        :rtype: str
        ---
        '''
        long_name = f'Manufacturing Year: { self.year }\nMaker: { self.make }\nModel: { self.model }.'
        return long_name

    def read_odometer(self) -> None:
        '''
        Provides the reading of the odometer.
        '''
        print(f'This car has { self.odometer_reading } miles on it.')

    def increment_odometer(self, miles) -> None:
        '''
        Increases the `odometer_reading` by the provided `miles`.
        ---
        :param miles: Distance of 1760 yards.
        :type miles: int
        :rtype: None
        ---
        '''
        self.odometer_reading += miles

#region Main

my_new_car = Car('audi', 'a4', 2019)
print(my_new_car.get_descriptive_name())
my_new_car.read_odometer()
my_new_car.odometer_reading = 50
my_new_car.read_odometer()

#region Requested-Code

Car.increment_odometer(my_new_car, 10)
Car.read_odometer(my_new_car)

Car.increment_odometer(my_new_car, 20)
Car.read_odometer(my_new_car)

#endregion

#endregion
```

# 9-4. Number Served:


```python
class Servings(Restaurant):
    '''
    Restaurant with an add-on feature of Number of Servings.
    '''

    def __init__(self) -> None:
        '''
        Initializing the class members' default values.
        '''
        self.number_served = 0

    def set_number_served(self, number_served) -> None:
        '''
        Sets the number of customers that have been served.
        ---
        :param number_served: New number of servings to be replaced.
        :type number_served: int
        :rtype: None
        ---
        '''
        self.number_served = number_served

    def increment_number_served(self, number_served):
        '''
        Increases the number of servings served to the customer.
        ---
        :param number_served: New number of servings to be altered.
        :type number_served: int
        :rtype: None
        ---
        '''
        self.number_served += number_served

    def describe_number_served(self) -> str:
        '''
        An additional method for `number_served` provides a string that decoratively describes the number of servings made for the instance.
        ---
        :return: A string demonstrating the number of servings.
        :rtype: str
        ---
        '''
        return (f'Number of Servings: { self.number_served }')
```

```python
#region Unit-Testing

if __name__ == '__main__':
    # Creating and initializing the object.
    servings = Servings()

    # Randomly iterating for setting the number of servings.
    print('set_number_served():')
    print('--------------------')
    for iteration in range(1, random.randint(1, 10 + 1)):
        Servings.set_number_served(servings, random.randint(1, 10 + 1))
        print(Servings.describe_number_served(servings))
    print('________________________\n')

    # Randomly iterating for updating the number of servings.
    print('increment_number_served():')
    print('--------------------------')
    for iteration in range(1, random.randint(1, 10 + 1)):
        Servings.increment_number_served(servings, random.randint(1, 10 + 1))
        print(Servings.describe_number_served(servings))    

#endregion
```

# 9-5. Login Attempts:


```python
class LoginAttempts(User):
    '''
    User with an add-on feature of providing login attempts.
    '''

    def __init__(self, login_attempts = 0) -> None:
        '''
        Initializes the default values for the class members.
        '''
        self.login_attempts = login_attempts
    
    def increment_login_attempts(self) -> None:
        '''
        Increases the login attempts by 1.
        '''
        self.login_attempts += 1

    def reset_login_attempts(self) -> None:
        '''
        Resets the login attempts to its default value. Currently, it is statically set to 0.
        '''
        self.login_attempts = 0

    def describe_login_attempt(self) -> str:
        '''
        Describes the `login-attempts` class member in a descriptive manner.
        ---
        :return: A decorated string describing the `login-attempts`.
        :rtype: str
        ---
        '''
        return (f'Login Attempts: { self.login_attempts }')
```

```python
#region Unit-Testing

# Initializing and creating instances of the class.
user = LoginAttempts()

# Increasing the login attempts.
user.increment_login_attempts()
print(LoginAttempts.describe_login_attempt(user))
user.increment_login_attempts()
print(LoginAttempts.describe_login_attempt(user))
user.increment_login_attempts()
print(LoginAttempts.describe_login_attempt(user))


# Reset the login attempts.
user.reset_login_attempts()
print(LoginAttempts.describe_login_attempt(user))

#endregion
```
