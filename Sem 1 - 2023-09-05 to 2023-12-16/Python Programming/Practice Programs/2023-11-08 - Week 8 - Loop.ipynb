# Question 5: Given a list iterate it and display numbers which are divisible by 5 and if you find number greater than 150 stop the loop iteration.
image.png

```python
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]
for num in list1:
    if num > 150:
        break
    elif num % 5 == 0:
        print(num)
```

# Question 10: Display Fibonacci Series up to 10 terms.
## Fibonacci Sequence: Each number is the sum of the two preceding ones, starting from 0 and 1.
image.png

```python
a, b = 0, 1
print("Fibonacci sequence:")
for _ in range(10):
    print(a, end=" ")
    a, b = b, a + b
```

# Question 11: Write a loop to find the factorial of any number.
<font color = '#FFA500'>NOTE: The factorial (symbol: !) means to multiply all whole numbers from our chosen number down to 1.</font>
image.png

```python
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print("Factorial of", num, "is:", factorial)
```

# Question 12: Python program to display all the prime numbers within a range.
## <font color = '#FFA500'>A prime number is a whole number that cannot be made by multiplying other whole numbers.</font>
### Example:
37 is a Prime Number because no other whole numbers multiply together to make it.
<br>
6 is not a Prime Number because it can be made by 2x3 = 6.
image.png

```python
start = 25
end = 40
print("Prime numbers between", start, "and", end, "are:")
for num in range(start, end + 1):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                break
        else:
            print(num)
```
